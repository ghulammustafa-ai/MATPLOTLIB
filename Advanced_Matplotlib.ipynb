### Annotations

```python
batters = pd.read_csv("/content/batter (1).csv")
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
plt.figure(figsize=(12,7))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s=sample_df["runs"])
plt.xlabel("AVERAGE")
plt.ylabel("STRIKE_RATE")
plt.title("Scatter Plot")
plt.axhline(130,color="red")
plt.axvline(25,color="red")
for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

### subplots

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(12,7))
ax[0].scatter(batters["avg"],batters["strike_rate"])
ax[1].scatter(batters["avg"],batters["runs"])
ax[0].set_xlabel("AVERAGE")
ax[0].set_ylabel("STRIKE_RATE")
ax[1].set_xlabel("AVERAGE")
ax[1].set_ylabel("RUNS")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(12,7))
ax[0,0].scatter(batters["avg"],batters["strike_rate"])
ax[0,1].scatter(batters["avg"],batters["runs"])
ax[1,0].hist(batters["avg"])
ax[1,1].hist(batters["runs"])
```

### 3d scatter plots

```python
fig = plt.figure()

ax = plt.subplot(projection="3d")

ax.scatter(batters["avg"],batters["strike_rate"],batters["runs"])
ax.set_xlabel("AVERAGE")
ax.set_ylabel("STRIKE_RATE")
ax.set_zlabel("RUNS")
```

### Surface plots

```python
x = np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx ,yy =np.meshgrid(x,y)
z=xx**2+yy**2
```

```python
fig = plt.figure(figsize=(12,7))
ax = plt.axes(projection="3d")
ax.plot_surface(xx,yy,z,cmap="viridis")
```

```python
z = np.sin(xx)+np.cos(yy)
```

```python
plt.figure(figsize=(12,7))
ax = plt.axes(projection="3d")
ax.plot_surface(xx,yy,z,cmap="viridis")
```

### countour plot

```python
fig = plt.figure(figsize=(12,7))
ax = plt.subplot()
ax.contour(xx,yy,z,cmap="viridis")
```

```python
fig = plt.figure(figsize=(12,7))
ax = plt.subplot()
ax.contourf(xx,yy,z,cmap="viridis")
```

### heatmap

```python
delivery = pd.read_csv("/content/IPL_Ball_by_Ball_2008_2022 (1).csv")
delivery.head()
```

```python
temp_df=delivery[(delivery["ballnumber"].isin([1,2,3,4,5,6])) & (delivery["batsman_run"] == 6)]
```

```python
grid = temp_df.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
```

```python
plt.figure(figsize=(12,7))
plt.imshow(grid)
plt.colorbar()
```
