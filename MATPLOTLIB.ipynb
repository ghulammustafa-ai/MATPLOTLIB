### 2D line plot


```python
 # bivariate analysis
 # numerical - numarical and categorical - categorical
 # time series data
```

```python
price = [45000,50000,55000,60000,15000]
year = [2015,2016,2017,2018,2019]

plt.plot(year,price)
```

```python
 batsman = pd.read_csv("/content/sharma-kohli.csv")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python

plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])

plt.title("Rohit sharma vs virat kohli")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
# color , linestyle, line width
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="dashed",linewidth=5) # also accepts color codes
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="dotted")

plt.title("Rohit sharma vs virat kohli")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="dashed",linewidth=5) # also accepts color codes
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="dotted",marker="d",markersize=5)

plt.title("Rohit sharma vs virat kohli")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
# label
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="dashed",linewidth=5,label="virat kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="dotted",marker="d",markersize=5,label="rohit sharma")

plt.title("Rohit sharma vs virat kohli")
plt.xlabel("season")
plt.ylabel("runs")
plt.legend(loc="best")
```

```python
price = [45000,50000,55000,60000,15000,150000000]
year = [2015,2016,2017,2018,2019,2021]

plt.plot(year,price)
plt.xlim(2017,2019)
plt.ylim(0,100000)
plt.grid()
```

### Bar chart

```python
 # bivariate uniavriate
 # numerical vs categorical
 # use acse : aggregate analysis of groups
```

```python
childrens = [10,20,30,40]
colors = ["red","green","blue","yellow"]

plt.bar(colors,childrens,color="black")
```

```python
plt.barh(colors,childrens,color="black")
```

```python
df = pd.read_csv("/content/batsman_season_record.csv")
```

```python
plt.bar(df["batsman"],df["2015"])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2 ,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0])  ,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2 ,df["2015"],width=0.2)

plt.xticks(np.arange(df.shape[0]),df["batsman"])

plt.show()
```

```python
childrens = [10,20,30,40]
colors = ["red red red red red","green green green green","blue blue blue blue blue","yellow yellow yellow yellow yellow"]
plt.plot(colors,childrens,color="black")
plt.xticks(rotation=90)
```

```python
# stacked bar chart
plt.bar(df["batsman"],df["2017"],label="2017")
plt.bar(df["batsman"],df["2016"],bottom=df["2017"],label="2016")
plt.bar(df["batsman"],df["2015"],bottom=(df["2016"])+df[("2017")],label="2015")

plt.legend()
```
